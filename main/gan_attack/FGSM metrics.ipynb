```python
import pickle
import os
from pprint import pprint
from matplotlib import pyplot as plt

from pylab import rcParams
rcParams['figure.figsize'] = 20, 10
```

```python
class Self:
    g_acc = None
    bb_acc = None
    d_acc = None
    bb_ong_acc = None
    run = None
    
    def __init__(self, metric_save_folder):
        self.run = metric_save_folder.split("\\")[-2]
        
        g_acc_file = os.path.join(metric_save_folder, 'g_acc.pkl')
        with open(g_acc_file, 'rb') as f:
            self.g_acc = pickle.load(f)

        bb_acc_file = os.path.join(metric_save_folder, 'bb_acc.pkl')
        with open(bb_acc_file, 'rb') as f:
            self.bb_acc = pickle.load(f)

        d_acc_file = os.path.join(metric_save_folder, 'd_acc.pkl')
        with open(d_acc_file, 'rb') as f:
            self.d_acc = pickle.load(f)

        bb_ong_acc_file = os.path.join(metric_save_folder, 'bb_ong_acc.pkl')
        with open(bb_ong_acc_file, 'rb') as f:
            self.bb_ong_acc = pickle.load(f)
            
    def __str__(self):
        return str(self.__dict__)
        
metrics = []
metric_parent = os.path.join('checkpoints', 'save', 'new_runs_metrics')
for root, dirs, files in os.walk(metric_parent, topdown=True):
    for d in dirs:
        if os.path.isfile(os.path.join(root, d, 'g_acc.pkl')):
            metric_save_folder = os.path.join(root, d)
            metrics.append(Self(metric_save_folder))
            
```

```python
xrange = min([len(m.g_acc) for m in metrics])

plt.figure()
plt.title("g_acc")
for m in metrics:
    plt.plot(list(range(xrange)), m.g_acc[-xrange:])
plt.legend([m.run for m in metrics])
plt.show()
```

```python
xrange = min([len(m.bb_acc) for m in metrics])

plt.figure()
plt.title("bb_acc")
for m in metrics:
    plt.plot(list(range(xrange)), m.bb_acc[-xrange:])
plt.legend([m.run for m in metrics])
plt.show()
```

```python
xrange = min([len(m.d_acc) for m in metrics])

plt.figure()
plt.title("d_acc")
for m in metrics:
    plt.plot(list(range(xrange)), m.d_acc[-xrange:])
plt.legend([m.run for m in metrics])
plt.show()
```

```python
xrange = min([len(m.bb_ong_acc) for m in metrics])

plt.figure()
plt.title("bb_ong_acc")
for m in metrics:
    plt.plot(list(range(xrange)), m.bb_ong_acc[-xrange:])
plt.legend([m.run for m in metrics])
plt.show()
```
